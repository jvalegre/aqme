#### Organic molecules with SMILES input, CSEARCH performs conformational sampling with RDKit, QDESCP optimizes with xTB and generates molecular descriptors with xTB and MORFEUS

###### Two columns titled "SMILES_SUFFIX" are provided. Descriptors are generated for the two components of each entry. For example, you might combine CCC with PC in the first entry, and generate descriptors independently for CCC and PC that will then be combined in the same array.

```python
from aqme.qdescp import qdescp

# set CSV input file
csv_file = 'CSV/two_smiles_cols.csv'

# run QDESCP descriptor generator, with:
# 1) Input CSV with SMILES to run (input=csv_file)
qdescp(input=csv_file)
```
