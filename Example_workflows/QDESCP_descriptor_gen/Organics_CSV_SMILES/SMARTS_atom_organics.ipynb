#### Organic molecules with SMILES input, CSEARCH performs conformational sampling with RDKit, QDESCP optimizes with xTB and generates molecular descriptors with xTB and MORFEUS

###### Two common and unique substructures (O and N atoms). By default, QDESCP would detect both types of atoms and would generate atomic descriptors of O and N. With the qdescp_atoms option and the "O" SMARTS pattern, we force the program to generate atomic descriptors only in O atoms. Molecular descriptors are always calculated as well.
###### Other SMARTS patterns can be used for atoms (O, P, N...) or functional groups (C=O, C#N...). That pattern must be repeated only once in the molecules.

```python
from aqme.qdescp import qdescp

# set CSV input file
csv_file = 'CSV/aminoalcohols.csv'

# run QDESCP descriptor generator, with:
# 1) Input CSV with SMILES to run (input=csv_file)
# 2) qdescp_atoms is used with a SMARTS pattern that represents O atoms (qdescp_atoms=['O'])
qdescp(input=csv_file,qdescp_atoms=['O'])
```
