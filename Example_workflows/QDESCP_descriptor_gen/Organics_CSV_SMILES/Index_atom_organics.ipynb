#### Organic molecules with SMILES input, CSEARCH performs conformational sampling with RDKit, QDESCP optimizes with xTB and generates molecular descriptors with xTB and MORFEUS

###### QDESCP will only generate descriptors for the atoms marked with the index 1 in the numbered SMILES. It is important to mark the target atoms in the SMILES (ie CC[P:1] to mark the P atom as the atom with index 1)

```python
from aqme.qdescp import qdescp

# set CSV input file
csv_file = 'CSV/phosphines_index.csv'

# run QDESCP descriptor generator, with:
# 1) Input CSV with SMILES to run (input=csv_file)
# 2) qdescp_atoms is used with the atom index 1 (qdescp_atoms=['1'])
qdescp(input=csv_file,qdescp_atoms=['1'])
```
