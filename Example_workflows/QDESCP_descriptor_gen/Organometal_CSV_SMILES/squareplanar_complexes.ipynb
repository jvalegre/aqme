#### Organometallic complexes with SMILES input, CSEARCH performs conformational sampling with RDKit, QDESCP optimizes with xTB and generates molecular descriptors with xTB and MORFEUS

###### Common and unique substructure (Pd, N and P atoms), so atomic Pd, N and P descriptors will be generated. Molecular descriptors are also calculated (by default).
###### Since the complexes are squareplanar (not default in AQME), we need to specify the squareplanar geometry for the complexes in the input CSV (column "complex_type"). Also, in this case we only accept complexes with the two Cl ligands in trans (specified in the "geom" column). Charges and multiplicity are included in the input CSV (required for metal complexes, "charge" and "mult" columns).

```python
from aqme.qdescp import qdescp

# set CSV input file
csv_file = 'CSV/Pd_complexes.csv'

# run QDESCP descriptor generator, with:
# 1) Input CSV with SMILES to run (input=csv_file)
qdescp(input=csv_file)
```
