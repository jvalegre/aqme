#### Organometallic complexes with SMILES input, CSEARCH performs conformational sampling with RDKit, QDESCP optimizes with xTB and generates molecular descriptors with xTB and MORFEUS

###### With qdescp_atoms and the "C=O" SMARTS pattern, we force the program to generate atomic descriptors only on the C and O atoms forming the carbonyl groups present in all structures. Molecular descriptors are also calculated (by default).
###### Other SMARTS patterns can be used for atoms (O, P, N...) or functional groups (C=C, C#N...). That pattern must be repeated only once in the molecules.
###### Since the complexes are tetrahedral (default in AQME), no need to specify any geometry for the complexes. Charges and multiplicity are included in the input CSV (required for metal complexes, "charge" and "mult" columns).

```python
from aqme.qdescp import qdescp

# set CSV input file
csv_file = 'CSV/CO-select_complexes.csv'

# run QDESCP descriptor generator, with:
# 1) Input CSV with SMILES to run (input=csv_file)
# 2) qdescp_atoms is used with a SMARTS pattern that represents carbonyl groups (qdescp_atoms=['C=O'])
qdescp(input=csv_file,qdescp_atoms=['C=O'])
```
